```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
data = pd.read_csv("data/train_transactions.csv")
fraud = pd.read_csv("data/train_fraudsters.csv")
users = pd.read_csv("data/train_users.csv")
```

```python
data.Fraud = data.Fraud.apply(lambda x: 0 if x== "0" else 1)
q = data.AMOUNT.quantile(0.99)
data = data[data["AMOUNT"] < q]
data.AMOUNT = data.AMOUNT/100
data.drop(["Unnamed: 0","STATE"],axis =1,inplace= True)
```

```python
users.PHONE_COUNTRY = users.PHONE_COUNTRY.apply(lambda x: x.split("||"))
users["phone"] = users.PHONE_COUNTRY.apply(lambda x: len(x))
users = pd.merge(users,fraud,how="left",left_on="ID",right_on = "user_id")
users["is_fraud"] = users.user_id.apply(lambda x:1 if x is np.nan  else 0)
```

```python
data["is_atm"] =data.MERCHANT_CATEGORY.apply(lambda x: 1 if x=="atm" else 0)
data["is_GBR"] = data.MERCHANT_COUNTRY.apply(lambda x: 1 if x=="GBR" else 0)
data["is_manu"] = data.ENTRY_METHOD.apply(lambda x: 1 if x=="manu" else 0)
data["is_bank_transfer"] = data.TYPE.apply(lambda x: 1 if x=="BANK_TRANSFER" else 0)
data["is_minos"] = data.SOURCE.apply(lambda x: 1 if x=="MINOS" else 0)
```

```python
need1 = data[["AMOUNT","USER_ID","is_atm","is_GBR","is_manu","is_bank_transfer","is_minos","Fraud"]]
need2 = users[["ID","phone"]]
```
